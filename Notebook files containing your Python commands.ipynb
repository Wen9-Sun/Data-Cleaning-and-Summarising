```python
#Calculate and output the outlier index row
import matplotlib.pyplot as plt
school.boxplot(column='Poorest (Wealth quintile)')
plt.show()
q1=school['Poorest (Wealth quintile)'].quantile(0.25)
q3=school['Poorest (Wealth quintile)'].quantile(0.75)
iqr=q3-q1
upper_threshold = q3 + 1.5 * iqr
lower_threshold = q1 - 1.5 * iqr
outliers_index = school[(school['Poorest (Wealth quintile)'] > upper_threshold) | (school['Poorest (Wealth quintile)'] < lower_threshold)].index
print(outliers_index)
```

```python
#Calculate the median to replace outliers
median_value = school['Poorest (Wealth quintile)'].median()
school.loc[outliers_index, 'Poorest (Wealth quintile)'] = median_value
```

```python
import matplotlib.pyplot as plt
school.boxplot(column='Richest (Wealth quintile)')
plt.show()
q1=school['Richest (Wealth quintile)'].quantile(0.25)
q3=school['Richest (Wealth quintile)'].quantile(0.75)
iqr=q3-q1
upper_threshold = q3 + 1.5 * iqr
lower_threshold = q1 - 1.5 * iqr
outliers_index = school[(school['Richest (Wealth quintile)'] > upper_threshold) | (school['Richest (Wealth quintile)'] < lower_threshold)].index
print(outliers_index)
```

```python
median_value = school['Richest (Wealth quintile)'].median()
school.loc[outliers_index, 'Richest (Wealth quintile)'] = median_value
```

## Identify and remove irrelevant observations (columns)

```python
school.drop(columns=['Region'], inplace=True)
school.drop(columns=['Sub-region'], inplace=True)
school.drop(columns=['Data source'], inplace=True)
school.drop(columns=['Time period'], inplace=True)
school
```

```python
#convert float to percentage
def convert_to_percentage(value):
    return '{:.0%}'.format(value)
school['Total'] = school['Total'].apply(convert_to_percentage)
school['Rural (Residence)'] = school['Rural (Residence)'].apply(convert_to_percentage)
school['Urban (Residence)'] = school['Urban (Residence)'].apply(convert_to_percentage)
school['Poorest (Wealth quintile)'] = school['Poorest (Wealth quintile)'].apply(convert_to_percentage)
school['Richest (Wealth quintile)'] = school['Richest (Wealth quintile)'].apply(convert_to_percentage)
school
```
